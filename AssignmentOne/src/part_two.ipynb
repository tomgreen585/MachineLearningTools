```python
#Creating Data and initializing seed
from sklearn.datasets import make_blobs
from sklearn.datasets import make_circles
from sklearn.datasets import make_classification

#Creating the data
xblob, yblob = make_blobs(n_samples=1000, n_features=2, random_state=42) #blob
xclass, yclass = make_classification(n_samples=1000, n_redundant=0, n_clusters_per_class=1, n_features=2, random_state=42) #class
xcirc, ycirc = make_circles(n_samples=1000, noise=0.3, random_state=42) #circ
```

```python
#KMeans Algorithm
from sklearn.cluster._kmeans import KMeans

kblobmeans = KMeans(n_clusters=3) #Blob
kblobmeans.fit(xblob)
kblobpred = kblobmeans.predict(xblob)
kclassmeans = KMeans(n_clusters=2) #Class
kclassmeans.fit(xclass)
kclasspred = kclassmeans.predict(xclass)
kcircmeans = KMeans(n_clusters=2) #Circles
kcircmeans.fit(xcirc)
kcircpred = kcircmeans.predict(xcirc)
```

```python
#AffinityPropagation Algorithm
from sklearn.cluster._affinity_propagation import AffinityPropagation

apblob = AffinityPropagation() #Blob
apblob.fit(xblob)
apblobpred = apblob.predict(xblob)
apclass = AffinityPropagation() #Class
apclass.fit(xcirc)
apclasspred = apclass.predict(xcirc)
apcirc = AffinityPropagation() #Circles
apcirc.fit(xcirc)
apcircpred = apcirc.predict(xcirc)
```

```python
#DBSCAN Algorithm
from sklearn.cluster._dbscan import DBSCAN

dbsblob = DBSCAN() #Blob
dbsblob.fit(xblob)
dbsOnepred = dbsblob.fit_predict(xblob)
dbsclass = DBSCAN() #Class
dbsclass.fit(xclass)
dbsTwopred = dbsclass.fit_predict(xclass)
dbscirc = DBSCAN() #Circles
dbscirc.fit(xcirc)
dbsThreepred = dbscirc.fit_predict(xcirc)
```

```python
#GuassianMixture Algorithm
from sklearn.mixture._gaussian_mixture import GaussianMixture

gmblob = GaussianMixture(n_components=3) #Blob
gmblob.fit(xblob)
gmOnepred = gmblob.predict(xblob)
gmclass = GaussianMixture(n_components=2) #Class
gmclass.fit(xclass)
gmTwopred = gmclass.predict(xclass)
gmcirc = GaussianMixture(n_components=2) #Circles
gmcirc.fit(xcirc)
gmThreepred = gmcirc.predict(xcirc)
```

```python
#Birch Algorithm
from sklearn.cluster._birch import Birch

brblob = Birch(n_clusters=3) #Blob
brblob.fit(xblob)
brOnepred = brblob.predict(xblob)
brclass = Birch(n_clusters=2) #Class
brclass.fit(xclass)
brTwopred = brclass.predict(xclass)
brcirc = Birch(n_clusters=2) #Circles
brcirc.fit(xcirc)
brThreepred = brcirc.predict(xcirc)
```

```python
#AgglomerativeClustering Alorithm
from sklearn.cluster._agglomerative import AgglomerativeClustering

acblob = AgglomerativeClustering(n_clusters=3) #Blob
acOnepred = acblob.fit_predict(xblob)
acclass = AgglomerativeClustering(n_clusters=2) #Class
acclass.fit(xclass)
acTwopred = acclass.fit_predict(xclass)
accirc = AgglomerativeClustering(n_clusters=2) #Circles
accirc.fit(xcirc)
acThreepred = accirc.fit_predict(xcirc)
```

```python
#MeanShift Algorithm
from sklearn.cluster._mean_shift import MeanShift

msblob = MeanShift() #Blob
msblob.fit(xblob)
msOnepred = msblob.predict(xblob)
msclass = MeanShift() #Class
msclass.fit(xclass)
msTwopred = msclass.predict(xclass)
mscirc = MeanShift() #Circles
mscirc.fit(xcirc)
msThreepred = mscirc.predict(xcirc)
```

```python
#Print on display
import matplotlib.pyplot as plt

result = {'kmeans':[kblobpred, kclasspred, kcircpred],
          'afprop':[apblobpred, apclasspred, apcircpred],
          'dbscan':[dbsOnepred, dbsTwopred, dbsThreepred],
          'gmmodel':[gmOnepred, gmTwopred, gmThreepred],
          'birch':[brOnepred, brTwopred, brThreepred],
          'aggcl':[acOnepred, acTwopred, acThreepred],
          'mshift':[msOnepred, msTwopred, msThreepred]}

data_name = ['make_blob', 'make_classification', 'make_circ']
algorithms = ['kmeans', 'afprop', 'dbscan', 'gmmodel', 'birch', 'aggcl', 'mshift']

fig, axs = plt.subplots(7, 3, figsize=(12, 21))
for i, algorithm in enumerate(algorithms):
    for j, data_point in enumerate(data_name):
        df = result[algorithm][j]
        ax = axs[i, j]
        if data_point == 'make_blob':
            data = xblob
        elif data_point == 'make_classification':
            data = xclass
        elif data_point == 'make_circ':
            data = xcirc
        ax.scatter(data[:, 0], data[:, 1], c=df)
        ax.set_title(algorithm + " - " + data_point)
plt.tight_layout()
plt.show()
```
